```python
df = pd.read_excel('Jan18.xlsx', sheet_name="MIS6")
df.head()
```

```python
#Visualization Set Up
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_context("paper")
sns.set_theme(style="ticks")
#Plot
sns.scatterplot(x="Age", y="GRuberd18O", hue="Author", data=df) 
plt.legend(title="Author",loc='lower right', fontsize=13)
plt.gca().invert_yaxis()

#Label
sns.despine(top= True, right= True)
plt.xlabel('Age [kyr BP]',fontsize=15, fontweight='bold')
plt.ylabel('18O [ PDB]',fontsize=15, fontweight='bold')
plt.title('G. ruber 18O During MIS 6',fontsize=20, fontweight='bold') 

plt.show()
```
